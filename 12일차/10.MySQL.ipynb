```python
SQL = '''
CREATE TABLE tempTable(
  id TINYINT NOT NULL AUTO_INCREMENT,
  user_name varchar(10) NOT NULL,
  phone varchar(30) NULL,
  PRIMARY KEY(id)
)
'''
print(SQL)
cursor = db.cursor()        # 커서 생성
cursor.execute(SQL)         # 커서에게 쿼리 실행 명령
cursor.close()              # 커서 제거
```

CHAR 타입과 VARCHAR 타입의 차이

CHAR(10)이나 VARCHAR(10)은 똑같이 10글자?
## 아니다!
CHAR(10)은 실제 입력 내용이 얼마던, 10자만큼의 메모리를 차지,
VARCHAR(10)는 10자가 안되면 그만큼의 메모리를 반환한다! -> 메모리 활용성이 더 좋음
그러나 속도는 CHAR가 더 빠르므로, 상황에 맞게 사용하는 것이 좋다
