```python
import bs4
import requests   # BS는 인터넷에서 자료를 가져오는 기능이 없으므로 request가 도와줌
```

```python
url = 'https://search.shopping.naver.com/search/category/100005307'
result = requests.get(url).text
result      # 노트북 검색 페이지의 모든 데이터를 가져옴
```

```python
bsObj = bs4.BeautifulSoup(result, 'html.parser')    # html 형태를 bs오브젝트로 파싱하여 저장
```

```python
items = bsObj.find_all('a',{'class': 'basicList_link__JLQJf'})  # a 태그 중에서 클래스명이 같은 것을 모두 찾아줘!
```

```python
print(items[0].text)        # 객체 내의 상위 4개 목록을 출력
print(items[1].text)
print(items[2].text)
print(items[3].text)
```
