```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,100)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x-0), color='blue')        # 색상의 이름으로 선의 색상 변경 가능
plt.plot(x, np.sin(x-1), color='g')           # 혹은 색상의 약어로 변경 가능 RGB, CMYK
plt.plot(x, np.sin(x-2), color='0.75')        # 0~1 사이의 회색조를 표현
plt.plot(x, np.sin(x-3), color='#FFDD44')     # 16진수로 표현한 RGB
plt.plot(x, np.sin(x-4), color=(1.0,0.2,0.3)) # RGB 값을 튜플로 표현
plt.plot(x, np.sin(x-5), color='chartreuse')  # 색상의 이름 직접 지정
```

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.k')
plt.plot(x, x+3, ':r')
```

```python
# 플론 조정하기: 축, 경계

plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)        # 시작점과 끝점을 역으로 지정하면 좌우 반전도 가능!
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')       # 그래프에 여백 적게 '타이트'하게 출력
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

```python
# 플롯에 레이블 붙이기

plt.plot(x, np.sin(x))
plt.title('A sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend()
```

```python
# 산점도의 출력

x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x,y,'o',color='black')   # 'o' 의 자리를 '마커' 라고 한다!
```

```python
rng = np.random.RandomState(0)
for marker in ['o', ',', 'x', '+', 'v', '<', '>', 's', 'd']:
  plt.plot(rng.rand(5), rng.rand(5), marker, label='marker={0}'.format(marker))
  plt.legend()
  plt.xlim(0, 1.8)
```

```python
from sklearn.datasets import load_iris # sckit-learn

iris = load_iris()
```

```python
# 오차 시각화
x = np.linspace(0,10,50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k')
```
