### 확률??

- 사건이 일어날 가능성

- 고전적 확률 정의
    (각 실험의 가능성이 모두 같을 경우)  
    실험 결과가 N가지 일 대, 일어난 사건 A가 m번 일어났으면 확률은
    
$$ m/N $$  

- 경험적 정의
    반복 횟수를 N이라 했을 때, 무한번 반복했을 때 발생하는 A의 횟수

$$ P(A) = \lim (m/N) $$


- 공리적 정의
    콜모코로프 공리적 정의 세가지
    
    $$ 0 \le P(A) \le 1 $$
    $$ P(S) = 1 $$

    실험이 표본공간의 부분집합이면서 다른 사건과 교집합이 없을 때,  
    각 실험의 확률을 더한 값과, 합집합의 확률은 서로 같다

    예를 들면 동전 두개를 던질 경우
    
    $$
    앞면 : H, 뒷면 : T
    $$
    $$
    표본공간 = S(HH, HT, TH, TT)
    $$
    앞이 안나오는 사건인 A(TT)와 뒷면이 안나오는 사건 B(HH)는 서로 교집합이 없다
    $$
    P(A\cup B) = P({TT,HH}) = P({TT}) + P({HH})
    $$
    $$
    P(A\cup B) = P(A) + P(B)
    $$

### 확률의 성질
- 공집합의 확률은 0
- 여집합의 확률
    실험 A의 여집합은
$$
1-P(A)
$$

- 포합-배제 원리
$$
P(A\cup B) = P(A) + P(B) - P(A\cap B)
$$

- 각 사건이 배반사상(event)일 경우

    $$
    P(A1 \cup A2 \cup A3 ... \cup An) = P(A1) + P(A2) + ... + p(An)
    $$

    A가 B의 부분집합일 때 $P(a) \le P(B)$

### 조건부 확률
- 어떤 사건 B가 발생했을 때($P(B) > 0$), 사건 A가 할생할 확률

$$ P(A|B) = \frac{P(A\cap B)}{P(B)} $$

- 사건 A가 발생했을 때, 사건 B가 발생할 확률

$$ P(B|A) = \frac{P(A\cap B)}{P(A)} $$

$$ P(A \cap B) = P(A|B)P(B)=P(B|A)P(A) $$

- 독립사건
$$ P(A|B) = P(A) $$
$$ P(B|A) = P(B) $$

### 베이즈 정리

- 분할사건  
    사건 K개를 모두 합집합 했을 때, S가 되는 경우  
    사건 K개가 일어났을 때, $A_1 \sim A_K$ 까지 모두가 서로 교집합이 없는 경우를 분할사건이라 한다  

- 사전 확률을 사후 확률을 통해서 추정하는 정리

$$ B = (B \cap {A_1}) \cup ... \cup (B \cap {A_K}) $$
$$ P(B) = P(B \cap {A_1} + ... + P(B \cap {A_K})) $$

$$ P({A_i}|B) = \frac{P(B|{A_i})P({A_i})}{P(B)} = \frac{P(B|{A_i})P({B_i})}{P(B|{A_1})P({A_1}) + P(B|{A_2})P(A_2) + ... + P(B|{A_K})P({A_K})}



### 확률 변수와 확률 분포

- 확률변수  
    표본 공간에서 정의된 실수값 함수  
    확률 변수들의 통계량을 찾고  
    확률 변수를 모수를 통해 추정  
    대문자 'X'로 표현

- 확률 분포  
    확률 변수 값과 확률을 대응시키는 표, 그래프, 함수

- 이산확률 변수  
    countably infinite 변수
    확률 질량 함수를 통해 확률을 얻을 수 있다.

    - 확률 질량 함수
    $$ P(a \le X \le B) =\sum f(x)$$

- 연속확률 변수  
하나로 특정할 수 없고, 셀 수 없는 변수  

    - 확률 밀도 함수  
        확률 밀도 함수 f(x)는 연속확률 변수 X에 대한 확률의 분포를 나타내는 함수  

        모든 x에 대해서 $f(x) \ge 0 $  
        확률 밀도 함수의 전체 면적은 1  
        특정 구간에서의 확률은 해당 구간의 확률밀도 함수를 적분한 값이다

### 정규 분포
- 정의  
    확률 변수 X의 평균이 m이고, 분산이 $\sigma$인 정규분포는 $X \sim N(m, \sigma)$

- 특징  
    모든 정규분포는 좌우 대칭  
    정규곡선의 위치는 평균에 의해서 결정된다.(평균이 중앙)  
    표본공간에 따라 평균과 표준편차는 다르지만, 평균으로부터 k배 표준편차 이내에 확률변수 X가 존재할 확률은 동일

### 표준 정규 분포

확률 변수를 정규화하여 표준편차가 1이고, 평균이 0인 형태의 정규 분포로 만든 것

### 분산과 표준편차

- 분산을 통해 확률분포 함수의 모양을 설명할 수 있다.
- 분산은 확률분포함수의 확률이 모여있는지, 퍼져있는지를 추정할 수 있는 수치적 파라미터이다.

$${\sigma^2} = Var[X] = E[(x-\mu)^2]$$

- 성질  
    $Var[X] \ge 0$  
    $Var[cX] = c^2Var[X]$

    $Var[X+Y] = Var[X] + Var[Y] + 2E[(X-{m_x})(Y-{m_y})]$
    * E는 기댓값(평균), $m_x, m_y$는 x,y의 평균    
<br/>

- 표본평균의 분산과 표본의 분산의 관계  
$$E[X] = E[\bar{X}]$$
$$Var[\bar{X}] = \frac{1}{N}Var[X]$$
