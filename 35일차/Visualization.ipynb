### 바 그래프

```python
data = {
    '바나나' : 3, '사과' : 2, '오렌지' : 10
}
print(data.keys(), data.values())

plt.bar(data.keys(), data.values())
plt.title('과일 개수')
plt.xticks([0,1,2], data.keys())
plt.yticks(sorted(data.values()))
plt.grid(True)
plt.xlabel('과일')
plt.ylabel('개수')
plt.show()
```

```python
# 가로 바 그래프 = barh
data = {
    '바나나' : 3, '사과' : 2, '오렌지' : 10
}

plt.barh([0,1,2], data.values())
plt.title('과일 개수')
plt.yticks([0,1,2], data.keys())
plt.xticks(sorted(data.values()))
plt.grid(True)
plt.ylabel('과일')
plt.xlabel('개수')
plt.show()
```

### 파이 차트

```python
ratio = [34,32,16,18]
labels = ['사과', '바나나', '멜론', '포도']
explode = [0.05,0.05,0.05,0.05]
colors = ['silver', 'gold', 'whitesmoke', 'lightgrey']
wedgeprops = {'width':0.7,'linewidth':5}

#파이차트 메서드 plt.pie
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=0, counterclock=False, shadow=True,
    explode=explode, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

### 박스 그래프 - 이상치를 찾는 그래프

```python
np.random.seed(0)

plt.style.use('default')

data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

fig, ax = plt.subplots()

ax.boxplot([data_a, data_b, data_c], notch=True, whis=1.5)
# whis = 수염 길이를 조정
ax.set_ylim(-10.0, 10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
plt.show()
```

```python
fig, ax = plt.subplots()

ax.boxplot([data_a, data_b, data_c], notch=True, whis=3, vert=True)
# whis = 수염 길이를 조정, vert = 가로,세로 조정
ax.set_ylim(-10.0, 10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
plt.show()
```

### 바이올린 그래프

```python
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

violin = plt.violinplot([data_a, data_b,data_c],
    showmeans=True, showextrema=True, showmedians=True)
# 평균값, 최대/최소값, 중간값 출력 = showmeans, showextrema, showmedians
plt.show()
```

```python
violin = plt.violinplot([data_a, data_b,data_c],
    showmeans=True, showextrema=True, showmedians=True)
plt.xticks(np.arange(1,4), labels=['A','B','C'])
violin['bodies'][0].set_facecolor('red')
violin['cbars'].set_edgecolor('red')
violin['cmaxes'].set_edgecolor('purple')
violin['cmins'].set_edgecolor('yellow')
violin['cmedians'].set_edgecolor('blue')
violin['cmeans'].set_edgecolor('green')
plt.show()
```
